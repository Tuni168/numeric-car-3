```python
from scipy.stats import boxcox
```

```python
skewed_data = np.random.chisquare(df=3, size=1000)
```

```python
transformed_data, lambda_value = boxcox(skewed_data)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(skewed_data, bins=30, color='blue', alpha=0.7)
plt.title('Before Box-Cox Transformation')
plt.subplot(1, 2, 2)
plt.hist(transformed_data, bins=30, color='green', alpha=0.7)
plt.title('After Box-Cox Transformation')
plt.show()
```

```python
print("Optimal lambda value:", lambda_value)
```
